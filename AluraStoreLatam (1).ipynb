### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

#1. Análisis de facturación



```python
ingreso1 = tienda1['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()

print("Ingreso Tienda 1:", ingreso1)
print("Ingreso Tienda 2:", ingreso2)
print("Ingreso Tienda 3:", ingreso3)
print("Ingreso Tienda 4:", ingreso4)
```

# 2. Ventas por categoría

```python
ventas_categoria1 = tienda1['Categoría del Producto'].value_counts()
ventas_categoria2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria4 = tienda4['Categoría del Producto'].value_counts()

import matplotlib.pyplot as plt

ventas_categoria1.plot(kind='bar', title='Tienda 1 - Ventas por Categoría')
plt.xlabel("Categoría")
plt.ylabel("Cantidad")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
reseñas = {
    'Tienda 1': tienda1['Calificación'].mean(),
    'Tienda 2': tienda2['Calificación'].mean(),
    'Tienda 3': tienda3['Calificación'].mean(),
    'Tienda 4': tienda4['Calificación'].mean()
}

plt.bar(reseñas.keys(), reseñas.values(), color='lightgreen')
plt.title("Promedio de Reseñas por Tienda")
plt.ylabel("Puntuación (0 a 5)")
plt.ylim(0, 5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos1 = tienda1['Producto'].value_counts()
producto_mas1 = productos1.idxmax()
producto_menos1 = productos1.idxmin()

productos2 = tienda2['Producto'].value_counts()
producto_mas2 = productos2.idxmax()
producto_menos2 = productos2.idxmin()

productos3 = tienda3['Producto'].value_counts()
producto_mas3 = productos3.idxmax()
producto_menos3 = productos3.idxmin()

productos4 = tienda4['Producto'].value_counts()
producto_mas4 = productos4.idxmax()
producto_menos4 = productos4.idxmin()





```

# 5. Envío promedio por tienda

```python
envio_prom1 = tienda1['Costo de envío'].mean()
envio_prom2 = tienda2['Costo de envío'].mean()
envio_prom3 = tienda3['Costo de envío'].mean()
envio_prom4 = tienda4['Costo de envío'].mean()

import matplotlib.pyplot as plt

# Crear diccionario con los valores promedios
envios_promedio = {
    'Tienda 1': envio_prom1,
    'Tienda 2': envio_prom2,
    'Tienda 3': envio_prom3,
    'Tienda 4': envio_prom4
}

# Crear el gráfico
plt.figure(figsize=(8,5))
plt.bar(envios_promedio.keys(), envios_promedio.values(), color='orange')
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío (USD)")
plt.xlabel("Tienda")
plt.ylim(0, max(envios_promedio.values()) * 1.2)  # para dar espacio arriba
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
